# Oral Support Notebook

## Elevator pitch

This project uses NLP on product descriptions to improve conversion on The North Face e-commerce website.
The 2 outputs are a simple recommender and a topic-based catalog reading.

## 5-minute structure

1. Business need
2. Dataset and preprocessing
3. Clustering and recommender
4. LSA and catalog insights
5. Limits and next steps

## Key messages

- TF-IDF is an interpretable baseline for product text.
- DBSCAN is relevant because the brief expects cosine distance and accepts outliers.
- Recommender MVP: same cluster + cosine similarity.
- LSA is useful for semantic navigation, not just recommendation.

## Typical jury questions

**Why DBSCAN?** It handles noise and non-spherical groups better than KMeans for this use case.

**Why is this not yet a production recommender?** Because there is no user behavior, only product content.

**What would you improve next?** Lemmatization, evaluation, metadata enrichment, and A/B testing.
